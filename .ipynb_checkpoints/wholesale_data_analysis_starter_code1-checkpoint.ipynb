## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df_cleanup = df[['unit_price','qty']].copy()
df_cleanup.loc[:,'line_subtotal'] = df_cleanup['unit_price']*df_cleanup['qty']
df_cleanup.head(3)

#Alternate method : Create a column that calculates the subtotal for each line using the unit_price and the qty
#def sub_total(row):
#    return row['unit_price'] * row['qty']

#df_cleanup['sub_total'] = df_cleanup.apply(sub_total, axis = 1)
#df_cleanup.head()
```

```python
df_shipping_price = df_cleanup[['unit_price','qty','line_subtotal']].copy()
df_shipping_price.head()
df_shipping_price['unit_weight'] = df[['unit_weight']].copy()
df_shipping_price['total_weight'] = df_shipping_price['unit_weight']*df_shipping_price['qty']
df_rearrange = df_shipping_price[['unit_price','unit_weight','qty','total_weight']]
df_rearrange.head(5)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

def shipping_price(row):
    shipping_price = row['total_weight'] * 10
    if (row['total_weight'] > 50):
        shipping_price = row['total_weight'] * 7
        return shipping_price
    return shipping_price

df_rearrange['shipping_price'] = df_rearrange.apply(shipping_price, axis = 1)
df_rearrange.head(3)

#Alternate method using lambda/apply.

#df_rearrange.loc[:,'shipping_price'] = df_rearrange['unit_weight'].apply(lambda i: i*7 if i >50 else i*10)
#df_rearrange.head(3)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
#df_cleanup('line_price') = df_cleanup['subtotal'] + df_cleanup['shipping_price'] + df_cleanup['sales_tax']
df_rearrange['line_subtotal'] = df_rearrange['unit_price']*df_rearrange['qty']
#df_rearrange.head()

df_rearrange['sales_tax'] = round(df_rearrange['shipping_price'] + df_rearrange['line_subtotal']) * 0.0925
#df_rearrange.head()

def line_price(row):
    total_price = row['shipping_price'] + row['line_subtotal'] + row['sales_tax']
    #return f"{row['shipping_price'] + row['subtotal']}"
    return f"{total_price:.2f}"

df_rearrange['line_price'] = df_rearrange.apply(line_price, axis=1)
df_rearrange.head(3)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df.loc[:,'line_cost'] = (df_rearrange.loc[:,'shipping_price']) + (df.loc[:,'unit_cost'] * df.loc[:,'qty']) 
df.head(3)
```

```python
# Create a column for the profit of each line using line cost and line price
df['line_profit'] = (df_rearrange['line_price'].astype('float') - df['line_cost'].astype('float'))
#df.head(3)
df['line_price'] = df_rearrange['line_price']
df.head(3)
```
